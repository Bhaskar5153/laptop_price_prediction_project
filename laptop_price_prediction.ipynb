```python
def cpu_preprocess(text):

    if text == 'Intel Core i5' or text == 'Intel Core i7' or text == 'Intel Core i3':
        return text
    else:
        if text.split()[0] == 'Intel':
            return 'other intel processor'
        else:
            return 'AMD processor'

```

```python
df['cpu_name'] = df['cpu_name'].apply(func=lambda x: cpu_preprocess(x))
```

Based on the analysis, you should consider removing the following features due to low correlation with Price or high multicollinearity:

Memory (low correlation with Price and high multicollinearity with other features)
Hybrid (very low correlation with Price)
HDD (multicollinearity with SSD and lower correlation with Price)
Ram (high correlation with Price)
PPI (moderate correlation with Price)
SSD (moderate correlation with Price)
Weight (weak correlation, but may add value in combination with other features)

Keep: Ram, PPI, SSD, and possibly Weight.
Remove: Memory, Hybrid, HDD, and Flash Storage.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
X_numeric_col = X.select_dtypes(exclude=['object']).columns
```

```python
X_numeric_col
```

```python
X_categorical_col = X.select_dtypes(include=['object']).columns
```

```python
preprocessor = ColumnTransformer(transformers=[
    ('num_col_transfomer', StandardScaler(), X_numeric_col),
    ('cat_col_transformer', OneHotEncoder(drop='first', sparse_output=True,handle_unknown='ignore'), X_categorical_col)
])
```

```python
preprocessor
```

```python
lr_model = LinearRegression()
```

```python
pipe = Pipeline(steps=[
    ('processor', preprocessor),
    ('lr_model', lr_model)
])
```

```python
pipe.fit(X_train, y_train)
```

```python
lr_pred = pipe.predict(X_test)
```

```python
# lr_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
r2_score(y_test, lr_pred)
```

```python
mean_absolute_error(y_test, lr_pred)
```

```python
mean_squared_error(y_test, lr_pred)
```

```python
lasso_model = Lasso(alpha=0.01)

lasso_pipe = Pipeline(steps=[
    ('processor', preprocessor),
    ('lasso_model', lasso_model)
])

lasso_pipe.fit(X_train, y_train)

lasso_pred = lasso_pipe.predict(X_test)

print(f'r2 score is : {r2_score(lasso_pred, y_test)}')
print(f'MAE: {mean_absolute_error(y_test, lasso_pred)}')
```
